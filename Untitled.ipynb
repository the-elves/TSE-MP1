```python
class Progress(git.remote.RemoteProgress):
    def update(self, op_code, cur_count, max_count=None, message=''):
        if max_count is None:
            per = 0
        else:
            per = cur_count / max_count
        
        num = int(per * 50)
        print ('[', end = '')
        for i in range(num):
            print('=', end='')
        print('>', end='')
        for i in range(50-num):
            print('-', end='')
        print(']                     \r', end='')
```

```python
f2url={}
for i,r in mcdf.iterrows():
    fc = r['files_changed'].split('<_**next**_>')
    for c in fc:
        cobj = json.loads(c)
        commregex = re.compile(r'/?([\w%]{40,43})/?')
        m = commregex.search(cobj['raw_url'])
        if m is None:
            continue
        fixed_commit = m.group(1)
        m = commregex.search(r['version_before_fix'])
        if m is None:
            continue
        vuln_commit = m.group(1)
        filename = cobj['filename'].split('/')[-1]
        if not ( filename.endswith('.c') or \
            filename.endswith('.cpp')):
            continue
        fixed_url = cobj['raw_url']
        vuln_url = commregex.sub('/'+vuln_commit+'/', fixed_url)
        f2url[filename] = (vuln_url, fixed_url, cobj['filename'])
print(f2url)
```

```python
os.chdir('/home/ajinkya/TSE/mp1')
print(os.getcwd())
errored_files = []
for f,u in tqdm(f2url.items()):
    url_before = u[0]
    url_after = u[1]
    try:
        if 'googlesource' in url_before:
#             urllib.request.urlretrieve(url_after+'?format=TEXT', filename = 'files/after/'+f)
#             after_root_url = url_before.split('%5E/')[0]+'%5E/'
            pass
        else:
            urllib.request.urlretrieve(url_before, filename =  'files/before/'+f)
            urllib.request.urlretrieve(url_after, filename = 'files/after/'+f)
    except:
        print(url_before)
        print(url_after)
        errored_files.append(f)
```
